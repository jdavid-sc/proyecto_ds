```python
rng = np.random.default_rng(42)

df = pd.DataFrame({
    "edad": rng.integers(18,65,100),
    "ingresos": rng.normal(2000,500,100),
    "gasto": rng.normal(1200,300,100)
})

df.head()
```

```python
print("shape:", df.shape)
df.describe()
```

```python
kpis = {
    "edad_promedio": df["edad"].mean(),
    "ingresos_promedio": df["ingresos"].mean(),
    "gasto_promedio": df["gasto"].mean()
}

kpis
```
